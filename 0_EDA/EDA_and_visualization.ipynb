# Farm labelled Dataset

## Sample of images per class

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

dataset_path = '../../dataset/0_raw_data/zenodo-Machine Learning Dataset for Poultry Diseases Diagnostics'

classes = ['cocci','healthy','ncd','salmo']

def display_sample_images(dataset_path, classes, samples_per_class=3):
    fig, axes = plt.subplots(len(classes), samples_per_class, figsize=(10, 2 * len(classes)))
    
    for i, cls in enumerate(classes):
        cls_path = os.path.join(dataset_path, cls, cls)
        image_files = os.listdir(cls_path)
        sample_files = random.sample(image_files, samples_per_class)
        
        for j, file in enumerate(sample_files):
            img_path = os.path.join(cls_path, file)
            img = Image.open(img_path)
            
            ax = axes[i, j]
            ax.imshow(img)
            ax.set_xlabel(file, fontsize=10)
            ax.set_xticks([])
            ax.set_yticks([])
    
    plt.tight_layout()
    plt.show()

display_sample_images(dataset_path, classes)
```

# Farm labelled Dataset

## Number of images for each class

```python
import os
import matplotlib.pyplot as plt

dataset_path = '../../dataset/0_raw_data/zenodo-Machine Learning Dataset for Poultry Diseases Diagnostics'
classes = ['cocci', 'healthy', 'ncd', 'salmo']

def count_images_per_class(dataset_path, classes):
    image_counts = {}
    
    for cls in classes:
        cls_path = os.path.join(dataset_path, cls, cls)
        image_files = [f for f in os.listdir(cls_path) if os.path.isfile(os.path.join(cls_path, f))]
        image_counts[cls] = len(image_files)
    
    return image_counts

image_counts = count_images_per_class(dataset_path, classes)

total_images = sum(image_counts.values())

image_percentages = {cls: (count / total_images) * 100 for cls, count in image_counts.items()}

for cls, count in image_counts.items():
    print(f'Class "{cls}" has {count} images, which is {image_percentages[cls]:.2f}% of the total.')

plt.figure(figsize=(10, 5))
bars = plt.bar(image_counts.keys(), image_counts.values(), color='skyblue')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.title('Number of Images per Class: Farm-labelled Dataset')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

plt.show()

plt.figure(figsize=(10, 5))
plt.pie(image_counts.values(), labels=image_counts.keys(), autopct='%1.1f%%', colors=['skyblue', 'orange', 'green', 'red'])
plt.title('Percentage of Images per Class: Farm-labelled Dataset')
plt.show()
```

# Completely Annotated and Resized Data

## Number of images per class

```python
import os
import matplotlib.pyplot as plt
from collections import defaultdict

dataset_path = '../../dataset/1_Object_detection/8_resized_640'
classes = ['cocci', 'healthy', 'ncd', 'salmo']

def count_images_per_class_in_single_folder(dataset_path, classes):
    image_counts = defaultdict(int)
    
    image_files = [f for f in os.listdir(dataset_path) if os.path.isfile(os.path.join(dataset_path, f)) and f.endswith('jpg')]
    
    for file in image_files:
        for cls in classes:
            if file.startswith(cls):
                image_counts[cls] += 1
                break  # Stop checking other classes once a match is found
    
    return image_counts

image_counts = count_images_per_class_in_single_folder(dataset_path, classes)

total_images = sum(image_counts.values())

image_percentages = {cls: (count / total_images) * 100 for cls, count in image_counts.items()}

for cls, count in image_counts.items():
    print(f'Class "{cls}" has {count} images, which is {image_percentages[cls]:.2f}% of the total.')

plt.figure(figsize=(10, 5))
bars = plt.bar(image_counts.keys(), image_counts.values(), color='skyblue')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.title('Total Number of Images per Class')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

plt.show()

plt.figure(figsize=(10, 5))
plt.pie(image_counts.values(), labels=image_counts.keys(), autopct='%1.1f%%', colors=['skyblue', 'orange', 'green', 'red'])
plt.title('Percentage of Images per Class')
plt.show()
```
