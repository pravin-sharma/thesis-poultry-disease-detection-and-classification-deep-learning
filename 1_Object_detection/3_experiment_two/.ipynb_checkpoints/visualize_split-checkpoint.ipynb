```python
import os
from collections import Counter
import matplotlib.pyplot as plt

# Define paths
base_dir = os.path.join(os.getcwd(), 'datasets/farm_labelled')
splits = ['train', 'test', 'valid']

# Function to extract class from filename (assuming class info is in the filename before the first dot)
def get_class(filename):
    return filename.split('.')[0]

# Dictionary to hold the counts per class per split
split_class_counts = {split: Counter() for split in splits}

# Iterate through each split and count the classes
for split in splits:
    images_dir = os.path.join(base_dir, split, 'images')
    if not os.path.exists(images_dir):
        print(f"Directory {images_dir} does not exist.")
        continue

    images = [f for f in os.listdir(images_dir) if f.endswith('.jpg')]
    classes = [get_class(img) for img in images]
    split_class_counts[split].update(classes)

# Visualize data count per class per split
def plot_data_distribution(split_class_counts):
    fig, ax = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
    fig.suptitle('Data Distribution per Class per Split')

    for i, split in enumerate(split_class_counts):
        counter = split_class_counts[split]
        ax[i].bar(counter.keys(), counter.values())
        ax[i].set_title(f'{split.capitalize()} Set')
        ax[i].set_xlabel('Class')
        ax[i].set_ylabel('Count')
        ax[i].set_xticks(range(len(counter)))
        ax[i].set_xticklabels(counter.keys())

    plt.show()

plot_data_distribution(split_class_counts)
```

```python
import os
from collections import Counter

# Define paths
base_dir = os.path.join(os.getcwd(), 'datasets/farm_labelled')
splits = ['train', 'test', 'valid']

# Function to extract class from filename (assuming class info is in the filename before the first dot)
def get_class(filename):
    return filename.split('.')[0]

# Dictionary to hold the counts per class per split
split_class_counts = {split: Counter() for split in splits}

# Iterate through each split and count the classes
for split in splits:
    images_dir = os.path.join(base_dir, split, 'images')
    if not os.path.exists(images_dir):
        print(f"Directory {images_dir} does not exist.")
        continue

    images = [f for f in os.listdir(images_dir) if f.endswith('.jpg')]
    classes = [get_class(img) for img in images]
    split_class_counts[split].update(classes)

# Print the counts
for split in splits:
    print(f'\n{split.capitalize()} Set:')
    for cls, count in split_class_counts[split].items():
        print(f'Class {cls}: {count} images')
```
